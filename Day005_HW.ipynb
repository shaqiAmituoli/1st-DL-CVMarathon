```python
import requests
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

import json
data = json.loads(response)
json_formatted = json.dumps(data, ensure_ascii=False,indent = 4)

print(json_formatted)
```

```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

# è³‡æ–™æ•¸é‡
idx = 0
for d in data:
    idx += 1

print("å…±æœ‰", idx, "ç­†è³‡æ–™ï¼")
    
#æ¬„ä½åç¨±
print("-----------")
print("æ¬„ä½åç¨±:\n")
for key in data[0]:
    print(key)
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

for d in data:
    print("æ¨™é¡Œ:", d['title'])
    print("è²¼æ–‡æ™‚é–“:", d['createdAt'])
    print("ç•™è¨€äººæ•¸:", d['commentCount'])
    print("æŒ‰è®šäººæ•¸:", d['likeCount'])
    print("------------------------------")
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
p = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
#print(p.text)
np = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
#print(np.text)

p_data = json.loads(p.text)
np_data = json.loads(np.text)

idx = 0
p_comment = 0
p_like = 0

for d in p_data:
    p_comment += d['commentCount']
    p_like += d['likeCount']
    idx += 1

print("ç†±é–€ - å…±å–å¾— ", idx, " ç­†è³‡æ–™")
print("ç†±é–€ç•™è¨€å¹³å‡:", '%.2f'%(p_comment/idx), " / ç†±é–€æŒ‰è®šå¹³å‡:", '%.2f'%(p_like/idx))
    
print("-----------------------------")
    
idx = 0
np_comment = 0
np_like = 0

for d in np_data:
    np_comment += d['commentCount']
    np_like += d['likeCount']
    idx += 1

print("éç†±é–€ - å…±å–å¾— ", idx, " ç­†è³‡æ–™")
print("éç†±é–€ç•™è¨€å¹³å‡:", '%.2f'%(np_comment/idx), " / éç†±é–€æŒ‰è®šå¹³å‡:", '%.2f'%(np_like/idx))
```
